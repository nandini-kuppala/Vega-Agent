```python
# AI Job Search Agent with Improved Error Handling and Mock Functionality
# This agent searches for relevant job postings based on user profile

import os
import json
import requests
import time
import random
from datetime import datetime
from typing import Dict, List, Any, Optional
from bs4 import BeautifulSoup
from urllib.parse import quote_plus

# Optional imports - handle gracefully if not available
try:
    from groq import Groq
    GROQ_AVAILABLE = True
except ImportError:
    GROQ_AVAILABLE = False
    print("Groq package not available. Using fallback mechanisms.")

class JobSearchAgent:
    def __init__(self, serpapi_key: str = None, groq_api_key: str = None):
        """Initialize the job search agent with API keys"""
        self.serpapi_key = serpapi_key
        self.groq_api_key = groq_api_key
        
        # Initialize Groq client if available and key is provided
        self.groq_client = None
        if GROQ_AVAILABLE and groq_api_key:
            try:
                self.groq_client = Groq(api_key=groq_api_key)
                print("GROQ client initialized successfully")
            except Exception as e:
                print(f"Error initializing GROQ client: {e}")
        
        self.search_results = []
        self.job_postings = []
        self.user_profile = None
        
        # Flag for using mock data when APIs fail
        self.use_mock_data = False
        
    def load_user_profile(self, profile_data: Dict[str, Any]):
        """Load and process the user profile"""
        self.user_profile = profile_data
        # Fix skill key if needed (handle both "skills" and "skill" keys)
        if "skills" not in self.user_profile and "skill" in self.user_profile:
            self.user_profile["skills"] = self.user_profile["skill"]
            
        # Clean up skills (fix typos, etc.)
        if "skills" in self.user_profile:
            cleaned_skills = []
            for skill in self.user_profile["skills"]:
                # Fix common typos
                if skill == "wed development":
                    skill = "web development"
                if skill == "app deelopment":
                    skill = "app development"
                cleaned_skills.append(skill)
            self.user_profile["skills"] = cleaned_skills
            
        print(f"Loaded profile for user with skills: {', '.join(profile_data.get('skills', []))}")
        
    def extract_search_keywords(self) -> Dict[str, Any]:
        """Extract relevant keywords from user profile for job searching"""
        if not self.user_profile:
            raise ValueError("No user profile loaded. Please load profile first.")
            
        # If GROQ client is available, use it to extract keywords
        if self.groq_client:
            try:
                # Use GROQ to extract relevant search terms from profile
                prompt = f"""
                Extract the most relevant job search keywords from this user profile:
                {json.dumps(self.user_profile, indent=2)}
                
                Return a JSON with these fields:
                1. skills: List of most important skills (max 5)
                2. job_titles: List of job titles to search for (max 3)
                3. location: String of location (just the city name)
                4. work_mode: Work mode preference (remote, hybrid, on-site)
                5. experience_level: Junior, mid, senior based on years
                """
                
                response = self.groq_client.chat.completions.create(
                    model="llama3-70b-8192",
                    messages=[{"role": "user", "content": prompt}],
                    response_format={"type": "json_object"}
                )
                
                try:
                    search_params = json.loads(response.choices[0].message.content)
                    print(f"Generated search parameters: {search_params}")
                    return search_params
                except json.JSONDecodeError:
                    print("Error parsing GROQ response. Using fallback parameters.")
            except Exception as e:
                print(f"Error using GROQ for keyword extraction: {e}")
        
        # Fallback parameters
        city = ""
        if "location" in self.user_profile and isinstance(self.user_profile["location"], dict):
            city = self.user_profile["location"].get("city", "")
        elif "location" in self.user_profile and isinstance(self.user_profile["location"], str):
            city = self.user_profile["location"]
            
        work_mode = "Remote"
        if "job_preferences" in self.user_profile and isinstance(self.user_profile["job_preferences"], dict):
            if "type" in self.user_profile["job_preferences"]:
                work_mode = self.user_profile["job_preferences"]["type"]
        
        experience = self.user_profile.get("experience_years", 0)
        if isinstance(experience, dict) and "$numberInt" in experience:
            experience = int(experience["$numberInt"])
        
        fallback_params = {
            "skills": self.user_profile.get("skills", [])[:5],
            "job_titles": ["Software Developer", "AI Engineer"],
            "location": city,
            "work_mode": work_mode,
            "experience_level": "Junior" if experience < 2 else "Mid-level"
        }
        
        # If user has previous job title, include it in job titles
        if "last_job" in self.user_profile and "title" in self.user_profile["last_job"]:
            fallback_params["job_titles"].insert(0, self.user_profile["last_job"]["title"])
            
        print(f"Using fallback search parameters: {fallback_params}")
        return fallback_params
    
    def search_web(self, search_params: Dict[str, Any], max_results: int = 10) -> List[Dict[str, Any]]:
        """Search the web for job postings using SERP API or fallback to mock data"""
        if self.use_mock_data:
            return self._get_mock_job_data("web", search_params, max_results)
            
        if not self.serpapi_key:
            print("No SERP API key provided. Using mock data.")
            self.use_mock_data = True
            return self._get_mock_job_data("web", search_params, max_results)
            
        skills_str = " ".join(search_params["skills"][:3])  # Use top 3 skills
        job_titles_str = " OR ".join([f'"{title}"' for title in search_params["job_titles"]])
        location = search_params["location"] if isinstance(search_params["location"], str) else search_params["location"].get("city", "")
        work_mode = search_params["work_mode"]
        
        # Create search queries
        queries = []
        for job_title in search_params["job_titles"]:
            query = f"{job_title} jobs {skills_str} {location} {work_mode}"
            queries.append(query)
        
        results = []
        for query in queries[:2]:  # Limit to first 2 queries to avoid rate limits
            encoded_query = quote_plus(query)
            url = f"https://serpapi.com/search.json?engine=google&q={encoded_query}&api_key={self.serpapi_key}&tbm=jobs"
            
            try:
                response = requests.get(url)
                response.raise_for_status()
                data = response.json()
                
                # Extract job listings
                if "jobs_results" in data:
                    results.extend(data["jobs_results"])
                    print(f"Found {len(data['jobs_results'])} jobs for query: {query}")
                
                # Respect API rate limits
                time.sleep(1)
                
            except Exception as e:
                print(f"Error searching web: {e}")
                self.use_mock_data = True
                return self._get_mock_job_data("web", search_params, max_results)
        
        # Deduplicate and limit results
        unique_results = []
        job_ids = set()
        for job in results:
            job_id = job.get("job_id", "") or job.get("title", "") + job.get("company_name", "")
            if job_id not in job_ids and len(unique_results) < max_results:
                job_ids.add(job_id)
                unique_results.append(job)
        
        self.search_results = unique_results
        return unique_results
    
    def search_linkedin(self, search_params: Dict[str, Any], max_results: int = 10) -> List[Dict[str, Any]]:
        """Search LinkedIn for job postings"""
        if self.use_mock_data:
            return self._get_mock_job_data("linkedin", search_params, max_results)
            
        if not self.serpapi_key:
            print("No SERP API key provided. Using mock data.")
            self.use_mock_data = True
            return self._get_mock_job_data("linkedin", search_params, max_results)
        
        job_titles = " OR ".join([f'"{title}"' for title in search_params["job_titles"]])
        location = search_params["location"] if isinstance(search_params["location"], str) else search_params["location"].get("city", "")
        skills = " ".join(search_params["skills"][:3])
        
        # Construct LinkedIn search query for SERP API
        query = f"{job_titles} {skills} {location}"
        encoded_query = quote_plus(query)
        url = f"https://serpapi.com/search.json?engine=linkedin_jobs&q={encoded_query}&api_key={self.serpapi_key}"
        
        try:
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()
            
            # Extract job listings
            results = []
            if "jobs_results" in data:
                results = data["jobs_results"][:max_results]
                print(f"Found {len(results)} jobs on LinkedIn")
            
            # Add source information
            for job in results:
                job["source"] = "LinkedIn"
            
            return results
            
        except Exception as e:
            print(f"Error searching LinkedIn: {e}")
            self.use_mock_data = True
            return self._get_mock_job_data("linkedin", search_params, max_results)
    
    def _get_mock_job_data(self, source: str, search_params: Dict[str, Any], count: int = 5) -> List[Dict[str, Any]]:
        """Generate mock job data when APIs fail"""
        print(f"Generating mock job data for {source}")
        
        job_titles = search_params["job_titles"]
        skills = search_params["skills"]
        experience_level = search_params["experience_level"]
        location = search_params["location"] if isinstance(search_params["location"], str) else search_params["location"].get("city", "")
        work_mode = search_params["work_mode"]
        
        companies = [
            "TechCorp", "AI Solutions", "DataSmart", "CodeNation", "PyRapid", 
            "JavaTech", "WebWizards", "AppGenius", "MLMasters", "DevDynamics"
        ]
        
        locations = [location, "Remote", "Bangalore", "Hyderabad", "Chennai", "Mumbai", "Delhi"]
        
        mock_jobs = []
        for i in range(count):
            job_title = random.choice(job_titles) if job_titles else f"{random.choice(['Junior', 'Entry-level'])} Developer"
            if "AI" in skills and random.choice([True, False]):
                job_title = f"AI {job_title}"
            
            company = random.choice(companies)
            
            job_location = random.choice(locations)
            if work_mode.lower() == "remote":
                job_location = "Remote" if random.random() > 0.3 else job_location
                
            description = f"We are looking for a {job_title} with skills in {', '.join(random.sample(skills, min(len(skills), 3)))}. "
            description += f"This is a great opportunity for {experience_level} developers to grow their career."
            
            salary_range = f"${random.randint(50, 120)}K - ${random.randint(120, 180)}K" if random.random() > 0.3 else ""
            
            mock_job = {
                "title": job_title,
                "company_name": company,
                "location": job_location,
                "via": source,
                "description": description,
                "detected_extensions": {
                    "posted_at": f"{random.randint(1, 30)} days ago",
                    "schedule_type": "Full-time"
                },
                "job_id": f"mock-{source}-{i}",
                "link": f"https://example.com/jobs/{i}",
                "source": source,
                "salary_info": salary_range
            }
            
            if random.random() > 0.7:
                mock_job["detected_extensions"]["salary_range"] = salary_range
                
            mock_jobs.append(mock_job)
            
        print(f"Generated {len(mock_jobs)} mock jobs from {source}")
        return mock_jobs
    
    def scrape_job_details(self, job_url: str) -> Dict[str, Any]:
        """Scrape or mock detailed job information"""
        if self.use_mock_data or job_url.startswith("https://example.com"):
            return {
                "full_description": "This position requires expertise in software development and machine learning. " +
                                  "The candidate will work on cutting-edge AI projects and be responsible for developing " +
                                  "and implementing algorithms. Requirements include proficiency in Python, experience with " +
                                  "ML frameworks, and good problem-solving skills.",
                "application_link": job_url,
                "salary_info": f"${random.randint(50, 120)}K - ${random.randint(120, 180)}K per year"
            }
        
        try:
            headers = {
                "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
            }
            response = requests.get(job_url, headers=headers, timeout=10)
            response.raise_for_status()
            
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Extract job details (implementation depends on target site structure)
            job_description = ""
            description_div = soup.find('div', class_=['description', 'job-description', 'jobDescriptionText'])
            if description_div:
                job_description = description_div.get_text(strip=True)
            
            # Extract application link
            apply_link = job_url
            apply_button = soup.find('a', string=lambda text: text and ('Apply' in text or 'application' in text.lower()))
            if apply_button and apply_button.get('href'):
                apply_link = apply_button['href']
            
            # Extract salary if available
            salary = ""
            salary_elem = soup.find(string=lambda text: text and ('$' in text or 'â‚¬' in text or 'Â£' in text) and 
                                  ('salary' in text.lower() or 'compensation' in text.lower()))
            if salary_elem:
                salary = salary_elem.strip()
            
            return {
                "full_description": job_description[:500] + "..." if len(job_description) > 500 else job_description,
                "application_link": apply_link,
                "salary_info": salary
            }
            
        except Exception as e:
            print(f"Error scraping job details: {e}")
            return {
                "full_description": "Failed to retrieve detailed description",
                "application_link": job_url,
                "salary_info": "Not available"
            }
    
    def filter_jobs(self, jobs: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
        """Filter jobs based on user profile and preferences"""
        if not self.user_profile or not jobs:
            return jobs
        
        # Use GROQ if available
        if self.groq_client and not self.use_mock_data:
            try:
                filtered_jobs = []
                
                # Process in batches to avoid context length issues
                batch_size = 5
                for i in range(0, len(jobs), batch_size):
                    batch = jobs[i:i+batch_size]
                    
                    # Create prompt for relevance scoring
                    prompt = f"""
                    User profile:
                    {json.dumps(self.user_profile, indent=2)}
                    
                    Job listings:
                    {json.dumps(batch, indent=2)}
                    
                    For each job, rate its relevance to the user profile on a scale of 1-10 and explain why.
                    Return a JSON array with objects containing: job_id, relevance_score, reason.
                    """
                    
                    response = self.groq_client.chat.completions.create(
                        model="llama3-70b-8192",
                        messages=[{"role": "user", "content": prompt}],
                        response_format={"type": "json_object"}
                    )
                    
                    relevance_data = json.loads(response.choices[0].message.content)
                    
                    # Add relevance scores to jobs and filter
                    for job in batch:
                        job_id = job.get("job_id", "") or job.get("title", "") + job.get("company_name", "")
                        for score_data in relevance_data.get("ratings", []):
                            if score_data.get("job_id") == job_id and score_data.get("relevance_score", 0) >= 6:
                                job["relevance_score"] = score_data.get("relevance_score")
                                job["relevance_reason"] = score_data.get("reason")
                                filtered_jobs.append(job)
                                break
                    
                    # Respect API rate limits
                    time.sleep(1)
                
                # Sort by relevance
                filtered_jobs.sort(key=lambda x: x.get("relevance_score", 0), reverse=True)
                return filtered_jobs
                
            except Exception as e:
                print(f"Error using GROQ for job filtering: {e}")
        
        # Fallback: simple keyword-based filtering
        print("Using fallback job filtering mechanism")
        user_skills = [s.lower() for s in self.user_profile.get("skills", [])]
        
        # Extract preferred roles
        preferred_roles = []
        if "job_preferences" in self.user_profile and "roles" in self.user_profile["job_preferences"]:
            preferred_roles = [r.lower() for r in self.user_profile["job_preferences"]["roles"]]
        
        # Check last job title
        last_job_title = ""
        if "last_job" in self.user_profile and "title" in self.user_profile["last_job"]:
            last_job_title = self.user_profile["last_job"]["title"].lower()
        
        # Filter based on simple keyword matching
        scored_jobs = []
        for job in jobs:
            score = 0
            title = job.get("title", "").lower()
            description = job.get("description", "").lower()
            company = job.get("company_name", "").lower()
            
            # Score based on skills
            for skill in user_skills:
                if skill.lower() in title:
                    score += 3
                if skill.lower() in description:
                    score += 1
            
            # Score based on preferred roles
            for role in preferred_roles:
                if role.lower() in title:
                    score += 2
                if role.lower() in description:
                    score += 1
            
            # Score based on similarity to last job
            if last_job_title and (last_job_title in title or title in last_job_title):
                score += 3
            
            # Add location and remote preferences
            if "location" in self.user_profile:
                user_location = ""
                if isinstance(self.user_profile["location"], dict):
                    user_location = self.user_profile["location"].get("city", "").lower()
                else:
                    user_location = str(self.user_profile["location"]).lower()
                
                job_location = job.get("location", "").lower()
                
                if user_location and user_location in job_location:
                    score += 2
                
                # Check if user prefers remote and job is remote
                work_mode = ""
                if isinstance(self.user_profile["location"], dict):
                    work_mode = self.user_profile["location"].get("work_mode", "").lower()
                elif "job_preferences" in self.user_profile:
                    work_mode = self.user_profile["job_preferences"].get("type", "").lower()
                
                if work_mode == "remote" and "remote" in job_location:
                    score += 3
            
            # Add relevance info to job
            job["relevance_score"] = score
            reason = f"This job matches {score//2} of your skills or preferences"
            job["relevance_reason"] = reason
            
            scored_jobs.append(job)
        
        # Filter and sort
        filtered_jobs = [job for job in scored_jobs if job["relevance_score"] >= 5]
        filtered_jobs.sort(key=lambda x: x["relevance_score"], reverse=True)
        
        return filtered_jobs
    
    def get_job_recommendations(self, max_results: int = 10) -> List[Dict[str, Any]]:
        """Get personalized job recommendations based on profile"""
        search_params = self.extract_search_keywords()
        
        # Search various sources
        web_results = self.search_web(search_params, max_results)
        linkedin_results = self.search_linkedin(search_params, max_results)
        
        # Combine and deduplicate results
        all_jobs = web_results + linkedin_results
        unique_jobs = {}
        for job in all_jobs:
            job_id = job.get("job_id", "") or job.get("title", "") + job.get("company_name", "")
            if job_id not in unique_jobs:
                unique_jobs[job_id] = job
        
        combined_jobs = list(unique_jobs.values())
        
        # Filter for relevance
        filtered_jobs = self.filter_jobs(combined_jobs)
        
        # Enrich with detailed information (limit to top results)
        enriched_jobs = []
        for job in filtered_jobs[:max_results]:
            if "link" in job and job["link"]:
                details = self.scrape_job_details(job["link"])
                job.update(details)
            enriched_jobs.append(job)
        
        self.job_postings = enriched_jobs
        return enriched_jobs
    
    def generate_job_report(self) -> str:
        """Generate a detailed report of job recommendations"""
        if not self.job_postings:
            return "No job postings found. Please run get_job_recommendations() first."
        
        if self.groq_client and not self.use_mock_data:
            try:
                prompt = f"""
                Create a detailed job recommendations report for the user with this profile:
                {json.dumps(self.user_profile, indent=2)}
                
                Based on these job listings found:
                {json.dumps(self.job_postings, indent=2)}
                
                The report should include:
                1. A personalized introduction
                2. Summary of search criteria
                3. Top job recommendations with why they're a good fit
                4. Next steps and application tips
                
                Format the report in Markdown for readability.
                """
                
                response = self.groq_client.chat.completions.create(
                    model="llama3-70b-8192", 
                    messages=[{"role": "user", "content": prompt}],
                    max_tokens=2000
                )
                
                return response.choices[0].message.content
                
            except Exception as e:
                print(f"Error generating report with GROQ: {e}")
        
        # Generate fallback report manually
        print("Using fallback report generation")
        
        # Extract user info for personalization
        skills = ', '.join(self.user_profile.get('skills', []))
        experience = self.user_profile.get('experience_years', 0)
        if isinstance(experience, dict) and "$numberInt" in experience:
            experience = int(experience["$numberInt"])
            
        location = ""
        if "location" in self.user_profile:
            if isinstance(self.user_profile["location"], dict):
                location = self.user_profile["location"].get("city", "")
            else:
                location = str(self.user_profile["location"])
        
        last_job = ""
        if "last_job" in self.user_profile and "title" in self.user_profile["last_job"]:
            last_job = self.user_profile["last_job"]["title"]
        
        report = "# Job Recommendations Report\n\n"
        
        # Introduction
        report += "## Personalized Introduction\n\n"
        report += f"Based on your profile as an {last_job or 'professional'} with {experience} year{'s' if experience != 1 else ''} of experience "
        report += f"and skills in {skills}, we've found several promising job opportunities "
        report += f"that match your preferences for {'remote work' if 'Remote' in str(self.user_profile) else 'work in ' + location}.\n\n"
        
        # Search criteria
        report += "## Search Criteria\n\n"
        report += f"* **Location**: {location or 'Any'}\n"
        report += f"* **Skills**: {skills}\n"
        report += f"* **Experience**: {experience} year{'s' if experience != 1 else ''}\n"
        report += f"* **Previous Role**: {last_job or 'Not specified'}\n"
        
        if "job_preferences" in self.user_profile:
            prefs = self.user_profile["job_preferences"]
            if "type" in prefs:
                report += f"* **Work Type**: {prefs['type']}\n"
            if "roles" in prefs:
                report += f"* **Preferred Roles**: {', '.join(prefs['roles'])}\n"
            if "short_term_goal" in prefs:
                report += f"* **Short-term Goal**: {prefs['short_term_goal']}\n"
        report += "\n"
        
        # Top recommendations
        report += "## Top Job Recommendations\n\n"
        
        for i, job in enumerate(self.job_postings[:5], 1):
            report += f"### {i}. {job.get('title', 'Job')} at {job.get('company_name', 'Company')}\n"
            report += f"* **Location**: {job.get('location', 'Not specified')}\n"
            
            # Add posting date if available
            posted_at = None
            if "detected_extensions" in job and "posted_at" in job["detected_extensions"]:
                posted_at = job["detected_extensions"]["posted_at"]
            elif "posted_at" in job:
                posted_at = job["posted_at"]
                
            if posted_at:
                report += f"* **Posted**: {posted_at}\n"
                
            # Add salary info if available
            salary = None
            if "salary_info" in job and job["salary_info"]:
                salary = job["salary_info"]
            elif "detected_extensions" in job and "salary_range" in job["detected_extensions"]:
                salary = job["detected_extensions"]["salary_range"]
                
            if salary:
                report += f"* **Salary**: {salary}\n"
                
            # Add relevance info
            if "relevance_score" in job:
                report += f"* **Match Score**: {job['relevance_score']}/10\n"
            if "relevance_reason" in job:
                report += f"* **Why It's a Good Fit**: {job['relevance_reason']}\n"
                
            # Add description snippet
            if "description" in job:
                desc = job["description"]
                if len(desc) > 200:
                    desc = desc[:200] + "..."
                report += f"* **Description**: {desc}\n"
            elif "full_description" in job:
                desc = job["full_description"]
                if len(desc) > 200:
                    desc = desc[:200] + "..."
                report += f"* **Description**: {desc}\n"
                
            # Add application link
            link = job.get("application_link", job.get("link", "#"))
            report += f"* **Apply**: [Apply Here]({link})\n\n"
        
        # Next steps and tips
        report += "## Next Steps and Application Tips\n\n"
        report += "### Application Tips\n\n"
        report += "1. **Tailor Your Resume**: Highlight your skills in " + ", ".join(self.user_profile.get('skills', [])[:3]) + " for each application.\n"
        report += "2. **Prepare Your Portfolio**: Showcase projects that demonstrate your expertise.\n"
        report += "3. **Research Companies**: Learn about their products, culture, and tech stack before interviews.\n"
        report += "4. **Highlight Experience**: Emphasize your experience as " + (last_job or "a developer") + ".\n\n"
        
        report += "### Interview Preparation\n\n"
        report += "* Practice technical questions related to " + ", ".join(self.user_profile.get('skills', [])[:3]) + "\n"
        report += "* Prepare to discuss your previous work at " + (self.user_profile.get("last_job", {}).get("company", "your previous company") if "last_job" in self.user_profile else "your previous company") + "\n"
        report += "* Be ready to explain your career goals, including " + (self.user_profile.get("job_preferences", {}).get("short_term_goal", "your short-term goals") if "job_preferences" in self.user_profile else "your short-term goals") + "\n\n"
        
        report += "Good luck with your job search! These opportunities align well with your skills and preferences. Consider applying to your top choices this week for the best results."
        
        return report

# Example usage
def main():
    # In a real application, get these from environment variables
    serpapi_key = "429bd9006140a06a57ae6aba8af6bd52f12be0c5"
    groq_api_key = "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"
    
    # Sample user profile (parsed from input)
    user_profile = {
        "user_id": "6809c002a03a7a1e240ab91e",
        "education": "Bachelor's Degree",
        "skills": ["Python", "Java", "AI", "NLP", "ML", "DL", "web development", "app development"],
        "current_status": "Looking for Work",
        "experience_years": 1,
        "last_job": {"title": "AI developer", "company": "OLVT"},
        "job_preferences": {
            "type": "Remote Work",
            "roles": ["Software"],
            "short_term_goal": "Upskill and crack good placement",
            "long_term_goal": "Yes, i want to be an entrepreneur"
        },
        "location": {"city": "Tirupati", "relocation": True, "work_mode": "Flexible"}
    }
    
    # Initialize and run the agent
    agent = JobSearchAgent(serpapi_key=serpapi_key, groq_api_key=groq_api_key)
    agent.load_user_profile(user_profile)
    
    # Get job recommendations
    jobs = agent.get_job_recommendations(max_results=10)
    print(jobs)
    
    # Generate and print report
    report = agent.generate_job_report()
    print(report)
    
    # Save results to file
    with open("job_recommendations.md", "w") as f:
        f.write(report)
    
    with open("job_data.json", "w") as f:
        json.dump(agent.job_postings, f, indent=2)


if __name__ == "__main__":
    main()
```
